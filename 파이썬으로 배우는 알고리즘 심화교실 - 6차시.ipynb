while 다음의 조건식이 참일 동안, 코드를 반복 실행합니다. 

**작동 방식**
* i가 0으로 설정된 후 while문으로 들어가 조건(i<10)을 만족하는지 살펴봅니다.
* i는 0으로 설정되었기 때문에 10보다 작으므로 조건을 만족합니다
* "Hello, world!"를 출력합니다.
* i를 1 증가합니다.
* 다시 조건식으로 돌아가 조건이 참인지 확인하고, 참이면 내부 코드를 실행합니다.
* 조건이 거짓이 될 때까지 반복합니다. 
* 그렇다면, i가 0에서 9까지 커지는 동안 Hello, world!를 10번 출력하게 됩니다. 

**초기식, 변화식, 조건식**
* 초기식과 변화식은 while문을 종료시키기 위한 식들입니다.  
* 예를 들어 변화식이 없다면, i는 계속 0이기 때문에 while문이 종료하지 않고 무한하게 반복됩니다.    
* 그러나 무한하게 반복되는 것은 현실적으로 불가능하고, 오류가 발생하기도 합니다.   
* 초기식과 변화식은 그런 오류를 방지하고 while문을 안정적으로 종료하기 위해 작성되는 코드입니다. 


**예제1 : while문으로 구구단 출력하기**

````python
# 예제_1_while문으로_구구단_출력하기 
n = int(input("몇 단을 출력해볼까요?"))
i= 1
while i<10:
    print(n,"*",i,"=",n*i)
    i = i+1  # i를 1~9까지 점점 커지도록 해줍니다.  
````

**예제2 : while문으로 반 평균 수학 점수 구하기**

````python
#예제_2_while문으로_반_평균_수학_점수_구하기
n = int(input("학생수를 입력하세요!"))
i = 0
sum = 0
while i<n:
    score = int(input("학생 "+ str(i+1) + "의 점수를 입력하세요"))  # 초기값을 0으로 설정했기 때문에 학생번호를 표현할 때 +1을 했습니다. (초기값을 1로 설정하는 방법도 있겠죠?)
    sum = sum + score 
    i = i+1
    
print("평균 점수 : ", sum/n)
````

### range 함수 활용하기
range() 함수는 아래와 같은 방법으로 사용됩니다. 

> **range(시작, 끝, 변화량)**  : 시작점 이상 끝점 미만의 범위에서, 변화량 만큼 변화 시킵니다.  
예를 들어 **range(0,10,1)** 코드는 0,1,2,...,9 를 나타냅니다.(끝점은 포함되지 않는다는 점에 주의해야 합니다.)   
**range(0,10,2)** 코드는 0, 2, 4, 6, 8 을 나타냅니다.(변화량이 2이기 때문에 2씩 변화하는 모습을 볼 수 있습니다.)   


> **range()안의 세 가지 값을 반드시 다 입력할 필요는 없습니다.**   
**range()안에 값이 두 개만 입력된 경우** : 두 값은 각각 시작점과 끝점을 의미하게 됩니다. 이때 변화량은 1로 자동 설정됩니다.   
**range()안에 값이 한 개만 입력된 경우** : 해당 값은 끝점을 의미하게 됩니다. 이때 시작점은 0, 변화량은 1로 자동 설정됩니다.    

> **변화량이 음수여도 상관 없습니다.**   
시작점에서 끝점으로 값이 증가되는 것이 아니라 변화량만큼 감소됩니다. 즉, 시작점이 끝점보다 큰 값이어야 합니다.

**예제 10 : for문과 if문을 함께 사용해 3의 배수만 더하기**


````python
#예제_10_for문과_if문을_함께_사용해_3의_배수만_더하기

n = input("정수를 입력하세요.") # 변수 n에 정수를 입력받자.
n = int(n) # 입력받은 n을 정수형으로 변환해준다.
sum = 0 # 총합을 저장할 변수 sum을 0으로 설정하자.

# 0에서 n까지 순서대로 i에 대입하고, i가 3의 배수일 때에만 sum에 더해주는 반복문입니다. 
for i in range(n+1): # range 함수는 n미만까지가 범위이므로, n+1까지 범위를 설정해주자. 
    if i % 3 == 0: # 3으로 나눈 나머지가 0이면, 3의 배수라는 뜻이죠? (% : 나머지 연산자)
        sum = sum+i  #  i가 3의 배수일 때에만(조건 성립 시에만) sum에 i를 더해줍니다.

print(sum)
````

### 반복문 2번 중첩해서 사용하기

for문 속에 for문을 넣으면 어떤 일이 일어날까요?<br>
for문 안에 또 다른 for문을 중첩해서 작성하는 코드를 "이중 for문"이라고 합니다.<br>
이중 for문을 사용하는 방법을 아래의 예제와 함께 알아봅시다.

**예제 11 : 이중 for문 사용해보기**


````python
# 예제_11_이중_for문_사용해보기

for i in range(4):
    for j in range(3):
        print("i=", i, ", j=", j)
    print("=============")
````

> for i in range(4) 와 같이 바깥에 있는 for문과 for j in range(3) 과 같이 안 쪽에 있는 for문으로 이루어져 있습니다.   
이중 for문의 수행 과정은 다음과 같습니다.   
바깥쪽 for문이 실행되어 데이터 하나를 i에 저장하면, 그때 안쪽 for문이 실행됩니다.   
안쪽 for문의 반복이 모두 끝날 때까지 i는 처음 저장된 그 데이터를 유지하고,   
안쪽 for문 반복이 끝난 후에야 바깥쪽 for문의 반복이 다시 수행되어 i의 데이터가 바뀝니다.   
실제로 위에 제시된 이중 for문을 보면 i는 j가 range(3)만큼의 반복을 모두 끝낼 때까지 변화하지 않습니다.     
그리고 i가 range(4)만큼의 반복을 하여 마지막 데이터인 3을 저장하고 있을 때,  
j가 반복을 모두 끝내면 이중 for문의 반복이 종료됩니다.   

#### 예제 11 해설
> ```python
> for i in range(4)  

>와 같이 바깥에 있는 for문과   

>```python  
> for j in range(3)  

> 과 같이 안 쪽에 있는 for문으로 이루어져 있습니다.   

> 바깥쪽 for문이 실행되어 데이터 하나를 i에 저장하면, 그때 안쪽 for문이 실행됩니다.   
안쪽 for문의 반복이 모두 끝날 때까지 i는 처음 저장된 그 값을 유지하고,   
안쪽 for문 반복이 전부 끝난 후에야 바깥쪽 for문이 다시 실행되어 i의 값이 바뀝니다.   

> i가 0일 때 j가 0에서 2까지 반복되고, 그 다음 i가 1일때 다시 j가 반복되며, 동일한 방식으로 i가 3, j가 2일 때까지 맨 안쪽 코드는 총 12번 반복 실행되고 코드가 종료됩니다.

**예제 12 : 이중 for문으로 패턴 그리기**

>(힌트) 위의 그림은 행이 10줄, 열은 10개의 숫자로 이루어져 있고, 각각은 바깥 for문과 안쪽 for문으로 구현됨에 유의하여 예제를 풀어보세요.


````python
#예제_12_이중_for문으로_패턴_그리기

for i in range(10):
    for j in range(1, 11):
        print(j, end=" ")
    print()
````

#### 예제 12 해설
> 이중 for문은 패턴 그리기 문제에 많이 활용됩니다.  
2차원 상의 패턴은 행과 열로 표현되는데, 바깥쪽 for문이 행, 안쪽 for문이 열을 나타내도록 코드를 구성하면 패턴을 쉽게 구현할 수 있습니다.

### break로 반복문 종료하기

반복문을 중간에 멈추고 싶다면, `break`를 사용할 수 있습니다.
* `break` : 반복문의 조건과 무관하게 반복문을 강제로 중단시켜줍니다. 
* 반복문으로 코드를 실행하던 중에 특정 조건에서 반복문을 종료하고 싶을 때 사용할 수 있습니다. 


**예제 8 : break 활용하기**

````python
#예제_8_break_활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break  # n%3==0 일 때 반복문을 종료합니다. 따라서 다음의 "n은 3의배수가 아닙니다!"은 출력되지 않습니다. 
    print("n은 3의배수가 아닙니다!")
````

### continue로 반복문 건너뛰기

반복문을 중간까지만 실행하고 다시 처음으로 돌아오고 싶다면? continue를 사용할 수 있습니다.

* continue 또한 break와 마찬가지로 반복문 내에서 사용됩니다.  
* `continue`는 반복문의 시작 부분으로 돌아가, for문의 변수를 진행하거나 while문의 반복조건을 판단하게 합니다.  
* 즉 반복 중에 continue를 만나면, continue 이후의 코드는 실행하지 않고, 다시 반복문의 조건 부분으로 돌아가서 조건이 부합하면 다음 반복을 이어서 수행합니다.   



**예제 9 : continue 활용하기**

````python
#예제_9_continue_활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break
    if n % 4 ==0:
        print("4의 배수가 입력되었습니다! 처음으로 돌아갑니다.")
        continue  # n%4==0 일 때 이후의 코드를 반복하지 않고, while문의 조건 부분으로 돌아갑니다. 
    print("n은 3의배수가 아닙니다!")
````

````python
# 정답코드

answer = "369"

history = []
result = []

for i in range(9):
    print(i, "번째 이닝입니다.(",i,"/9)")
    player = input("정답을 입력해주세요")
    print("입력한 정답은 ", player, " 입니다.")
    history.append(player)

    ### 볼, 스트라이크 판별
    s = 0
    b = 0
    for j in range(len(player)):
        if player[j] == answer[j]:
            s += 1
        elif player[j] in answer:
            b += 1
    
    result.append([s,b])

    ### 정답, 아웃 판정
    if s == 3:
        print("정답입니다.")
        break
    elif s == 0 and b == 0:
        print("아웃입니다.")
    else:
        print(s, " 스트라이크, ", b," 볼 입니다.")

    print("지금까지의 기록입니다 : ", history, result)
    print("-"*50)
````
